```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis, skew

np.random.seed(34)

normal = np.random.normal(size = 100)

#Aprovecho el tiempo extra para plotearlo

plt.hist(normal, bins=16, color='blue')

plt.title('Distribución')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()

#Y sus principales métricas

media = np.mean(normal)
mediana = np.median(normal)
desviacion_estandar = np.std(normal)
varianza = np.var(normal)
curtosis = kurtosis(normal)
asimetria = skew(normal)

print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Desviación estándar: {desviacion_estandar}")
print(f"Varianza: {varianza}")
print(f"Curtosis: {curtosis}")
print(f"Asimetría: {asimetria}")
```

```python
#Lo mismo para el de chi^2

chi = np.random.chisquare(3, 100)

plt.hist(chi, bins=16, color='blue')

plt.title('Distribución')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()

#De nuevo con sus métricas

media = np.mean(chi)
mediana = np.median(chi)
desviacion_estandar = np.std(chi)
varianza = np.var(chi)
curtosis = kurtosis(chi)
asimetria = skew(chi)

print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Desviación Estándar: {desviacion_estandar}")
print(f"Varianza: {varianza}")
print(f"Curtosis: {curtosis}")
print(f"Asimetría: {asimetria}")


```

```python
#Ej 2

#Programa que calcule la desviación estándar en python de esta tabla


tabla = [4, 2, 5, 8, 6]

media = sum(tabla) / len(tabla)

#En lugar se sacar el bucle for, se puede hacer así en una sola línea
suma_cuadrados_diferencias = sum((x - media) ** 2 for x in tabla)

varianza = suma_cuadrados_diferencias / (len(tabla) - 1)

desviacion_estandar = varianza ** 0.5

print(f"Los datos de la tabla tienen una desviación estándar de {desviacion_estandar}")







```
